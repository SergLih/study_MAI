```python
import matplotlib
import numpy as np
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt
import seaborn as sns
# Define our surface

delta = 0.025
l1 = 4
l2 = 5
X = np.arange(0, l1+delta, delta)
Y = np.arange(0, l2+delta, delta)
x, y = np.meshgrid(X, Y)

z = np.sin((np.pi / l1) * x) * (-np.exp((-np.pi / l1) * y) + np.exp((np.pi / l1) * y)) / ((np.pi / l1) * (np.exp((np.pi / l1) * l2) + np.exp((-np.pi / l1) * l2)))

sns.set(font_scale=1.1)
fig, ax = plt.subplots(figsize=(l1, l2))
CS = ax.contour(x, y, z)
# ax.clabel(CS, CS.levels, inline=True, fmt='%0.2f', fontsize=10)
for i, label in enumerate(CS.cvalues): 
    CS.collections[i].set_label(np.round(label, 1))
ax.legend(loc='lower right')
ax.set_xlabel('x')
ax.set_ylabel('y')
#$\frac{1}{4}^{\prime\prime}$
#plt.title(r'$\frac{pi}{5}$')
#fig = plt.figure()
#fig.suptitle()

plt.title(r'$U(x, y) = \sin(\frac{\pi}{4} * x)* (-e^{\frac{-\pi}{4}*y} + e^{\frac{\pi}{4}*y})\,\,/\,\,(\frac{\pi}{4}*(e^{\frac{\pi}{4}*5}+e^{\frac{-\pi}{4}*5}))$')
plt.show()
```

```python
import matplotlib
import numpy as np
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.integrate as quad

delta = 0.025
l = 0.1
n_iter = np.linspace(1, 1000, 1000)
X = np.arange(0, l+delta, delta)
T = np.arange(0, 1000, delta)
x, t = np.meshgrid(X, T)
```

```python
def my_sum1(x, t, a=10**-6, 
           eps=1e-7, max_iter=10**3):
    def lambda_n(n):
        return ((2*n+1)*np.pi)/2*l
    
    def pi_n(n):
        return np.pi * (1 + 2*n)
    
    s1 = 0
    n = 0
    for n in range(0, max_iter):
        s_prev = s1
        s1 += 32*a/np.pi**4 * 1 / (1 + 2 * n)**4 * \
                (1000*(-1)**n * (1 - np.cos(5000*pi_n(n)*t)) - \
                 np.sin(5000 * pi_n(n) * t) + 5000*pi_n(n)*t) * \
                np.sin(5 * pi_n(n) * x)

        if np.abs(s1 - s_prev) < eps:
            break

    return s1/10**6
```

```python
my_sum1(0.1, 1000, a=1e-6)
```

```python
delta1 = 0.005
delta2 = 3
l = 0.1
X = np.arange(0, l+delta1, delta1)
T = np.arange(0, 500, delta2)
Z = []
for x in X:
    print('.', end='')
    zz = []
    for t in T:
        zz.append(my_sum1(x, t, a=1e-6))
    Z.append(zz)
```

```python
sns.set(font_scale=1.1)
fig, ax = plt.subplots(4, 2, figsize=(8, 12), sharey = True)
i = -1
for t in [0, 5, 10, 50, 100, 500, 800, 1000]:
    i += 1
    U = []
    X = np.linspace(0, 0.1, 100)
    for x in X:
        U.append(my_sum1(x, t, a=1e-6))
    ax[i // 2, i%2].plot(X, U)
    ax[i // 2, i%2].set_title("t = " + str(t))
#     ax.set_ylim((-0.001, 3))

plt.tight_layout()
plt.show()
    
    
```
