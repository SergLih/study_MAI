```python
class BiCG:
    def __init__(self, slau, eps=1e-5, max_iter = 1000):
        self.matrix = slau[0]
        self.b = slau[1]
        self.eps = eps
        self.order = len(self.matrix)
        self.max_iter = max_iter
        # if x0 is None else x0

#Википедия: Стабилизированный метод бисопряжённых градиентов        
    def BiCGStab(self):
        x0 = np.zeros_like(slau[1])
        r0 = self.b - self.matrix @ x0
        r1 = r0
        rho0, alpha0, omega0 = 1, 1, 1
        v0, p0 = np.zeros_like(self.b), np.zeros_like(self.b)
        k = 0
        while True:
            rho = r1 @ r0
            beta = (rho * alpha0) / (rho0 * omega0)
            p = r0 + beta * (p0 - omega0 * v0)
            v = self.matrix @ p
            alpha = rho / (r1 @ v)
            s = r0 - alpha * v
            t = self.matrix @ s
            omega = (t @ s) / (t @ t)
            x = x0 + omega * s + alpha * p
            r = s - omega * t
            k += 1
            if norm(r) < self.eps:
                break
            r0 = r
            rho0 = rho
            alpha0 = alpha
            omega0 = omega
            v0 = v
            p0 = p
            x0 = x
        return x
    
    def bicg(self):
        x = np.ones_like(slau[1])
        r = self.b - self.matrix @ x
        rs, p, ps = r[:], r[:], r[:]
#         rs = np.ones_like(r)
        #print(r, rs)
        assert r @ rs != 0
        

        #matrix_tmp = np.array(self.matrix)
        #b = a.transpose()
        #print(self.matrix)
#         matrix_transpose = [[self.matrix[row][col] for row in range(0, self.order) ] for col in range(0, self.order) ]
#         matrix_transpose = np.array(matrix_transpose)
        #print(matrix_transpose)
#         rho, z, s = r, r, r
#         p, ps = r, rs
        k = 0
        while k < self.max_iter:
            x_prev, r_prev, rs_prev = x[:], r[:], rs[:]
#             Ap = self.matrix @ p
            alpha = np.dot(r, rs) / np.dot((self.matrix @ p), ps)
            print(alpha)
            x += alpha * p
            r -= alpha * self.matrix @ p
            rs -= alpha * self.matrix.T @ ps #matrix_transpose * s0
            beta = np.dot(r, rs) / np.dot(r_prev, rs_prev)
            if np.abs(beta) < self.eps or norm(r) < self.eps:
                break
            p = r + beta*p
            ps = rs + beta*ps
            print(x.round(5))
            k += 1
            
        print(r @ rs)
        print(self.matrix @ p @ ps)
        return x
    
    def precond_solve(self):
        k_mtrx = spilu(self.matrix)
        x0 = np.zeros_like(slau[1])
        r0 = self.b - self.matrix @ x0
        r2 = r0
        p = r0
        k = 0
        while True:
            if r0 @ r2 == 0:
                exit()
            pp = k_mtrx.solve(p)
            tmp = self.matrix @ pp
            alpha = (r0 @ r2) / (tmp @ r2)
            s = r0 - alpha * tmp
            if norm(s) < self.eps:
                k += 1
                x = x0 + alpha * pp
                break
            z = k_mtrx.solve(s)
            tmp2 = self.matrix @ z
            omega = (tmp2 @ s) / (tmp2 @ tmp2)
            x = x0 + alpha * pp + omega * z
            r = s - omega * tmp2
            beta = (r @ r2) / (r0 @ r2) * (alpha / omega)
            p = r + beta * (p - omega * tmp)
            k += 1
            if norm(r) < self.eps:
                break
            x0 = x
            r0 = r
        return x
    
```
